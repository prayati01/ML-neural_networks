```python
import matplotlib.pyplot as plt #we are not importing whole lib just impoting the pyplot modeule of it
```

```python
x = [1,2,3,4,5] # consider these as pts on x axis
y = [1,2,3,4,5] #pts on y axis

plt.scatter(x,y) # we will scatter the points on the graph
plt.plot(x,y)
plt.show() #use to show the graph

#now it has just scattered th eppoint sbut not plotted them and to plot them , we need to give the cmd plt.plot(x,y)
```

```python
#when we need to plot multiple things
x = [1,2,3,4,5] # consider these as pts on x axis
x2= [1,2,3,4,5]
y = [1,2,3,4,5] #pts on y axis
y2 =[1,2,3,4,5]
y2.reverse()

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
#when u want ot vhange olors
x = [1,2,3,4,5] # consider these as pts on x axis
x2= [1,2,3,4,5]
y = [1,2,3,4,5] #pts on y axis
y2 =[1,2,3,4,5]
y2.reverse()

plt.plot(x,y, "r") # the color of line will be chnaged to red,  to make a red colored line in the form of dash use "r--",
plt.plot(x2,y2)
plt.show()
```

```python
#to plot a equation
import numpy as np

x= np.array([1,2,3,4,5,5])
y = 2*(x**4) +100
plt.plot(x,y)
plt.show()
#the graph is plotted but is not unique
```

```python
# to make that graph unique, that is the grph in o/p willl be smooth
x=np.arange(0,10,0.1)
y = 2*(x**4)+100

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,10,0.25)
y = x**4
plt.plot(x,y, color="red", marker = "p", linewidth =2)
# if u want the points to be pointed as pentagon , ue marker  = p, to chng width of line
plt.show()
```

```python
x= np.arange(0,10,0.25)
y = x**4
y2 = x**3
plt.plot(x,y, color="red" , label ="x**4")
plt.plot(x,y2, color="blue", label = "x**3")
plt.xlabel("x", fontsize =15)
plt.ylabel("y", fontsize =15)
plt.legend()# funciton used to show the label
plt.title("Chart")
plt.axis([0,20,0,10000])
plt.grid()
plt.show()
```
